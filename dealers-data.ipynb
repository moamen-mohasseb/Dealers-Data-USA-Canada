```python
driver= webdriver.Chrome('c:/WebDriver/bin/chromedriver.exe')
driver.get("https://www.360yieldcenter.com/dealer-locator/")
time.sleep(2)
for state in states:
    elem = driver.find_element_by_id('wpsl-search-input')
    elem.clear()
    elem.send_keys(state)
    
    #elem = driver.find_element_by_name('ctl00$ctl54$g_358d628f_a927_4218_9168_f21c190735a7$ctl00$ddlSpeciality')
    #elem.send_keys(spec)
    elem = driver.find_element_by_id('wpsl-search-btn')
    elem.click()
    time.sleep(4)
    try:
        #WebDriverWait(driver, 10).until(EC.frame_to_be_available_and_switch_to_it((By.ID,"ptifrmtgtframe")))
        WebDriverWait(driver, 6).until(EC.visibility_of_element_located((By.XPATH, "//span[@id='wpsl-street']")))
        print('before________________________________')
    except:     
        print('after________________________________')
        s=driver.page_source
        soup=bs(s,'html')
        tags=soup.find_all('div',"wpsl-store-location")
        tags    
        for s in tags:
            tag=s.text.strip().replace('\n\n','\n')
               # print(tag)
            store_name=tag.strip().split('\n')[0].replace('\n',' ')
            try:
                store_address=tag.strip().split('\n')[1].replace('\n',' ')+ '\n'+tag.strip().split('\n')[2].replace('\n',' ')
            except:
                store_address=tag.strip().split('\n')[1].replace('\n',' ')
            try:
                store_phone=re.findall('Phone:.+\d+.\d+.\d+',tag)[0]
                store_phone=store_phone[6:].replace('\n','')
            except:
                store_phone=''
            try:
                store_email=re.findall('Email:.[A-Za-z0-9@\._\-]+',tag)[0]
                store_email=store_email[7:].replace('\n','')
            except:
                store_email=""
            try:
                store_fax=re.findall('Fax:.\d+.\d+.\d+',tag)[0]
                store_fax=store_fax[4:].replace('\n','')
            except:
                store_fax=''
            dict_data={'State':state,'store_name':store_name,'store_address':store_address,'store_phone':store_phone,'store_email':store_email,'store_fax':store_fax}
            store_array.append( dict_data)
            print(dict_data,'\n')
    continue
print( store_array,'\n')
```
